```python
import cv2
import numpy as np

#video capture
video = cv2.VideoCapture(0)

```

```python
try:
    while True:

        ret, frame = video.read()

        if not ret:
            print("Failed to capture video")
            break

        cv2.imshow("frame", frame)

        b = frame[:, :, 0]
        g = frame[:, :, 1]
        r = frame[:, :, 2]

        #computing mean
        b_mean = np.mean(b)
        g_mean = np.mean(g)
        r_mean = np.mean(r)

        if b_mean > g_mean and b_mean > r_mean:
            print("The image is predominantly blue")

        elif g_mean > b_mean and g_mean > r_mean:
            print("The image is predominantly green")

        else:
            print("The image is predominantly red")

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    video.release()
    cv2.destroyAllWindows()
    
```
